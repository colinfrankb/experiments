```python
import numpy as np
import pandas as pd
import pandas_profiling
from pandas_profiling import ProfileReport
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
service_data_df = pd.read_csv("Serviceproviderpercitydata_modified_2021-4-21_1126.csv")

service_data_jhb_df = pd.read_csv("Serviceproviderpercitydata_modified_2021-4-21_1155.csv")
```

```python
service_data_df
```

```python
service_data_jhb_df
```

```python
service_data_ct_prof = ProfileReport(service_data_df, title='Service data for Cape Town', explorative = True)

service_data_ct_prof
```

```python
service_data_df = service_data_df[['service_name', 'connection_id']]

service_data_jhb_df = service_data_jhb_df[['service_name', 'connection_id']]
```

```python
service_data = service_data_df.groupby(['service_name']).count()

service_data
```

```python
service_data_jhb = service_data_jhb_df.groupby(['service_name']).count()

service_data_jhb

```

```python
plot = service_data.plot.pie(y='connection_id', figsize=(10, 10))
```

```python
plot = service_data_jhb.plot.pie(y='connection_id', figsize=(10, 10))
```

```python
plot = service_data.plot.pie(subplots=True, figsize=(11, 6))
```

```python
ax = service_data.plot.bar(stacked=True)
```

```python
ax = service_data_jhb.plot.bar(stacked=True)
```
